```python
from pyexpat import model
import pandas as pd
import seaborn as sns
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv(
    '/home/huigang/codespace/Rehabilitation_Assessment/software/data/dataset/康复评估动作/dataset/allData.csv')
dataTest = pd.read_csv(
    '/home/huigang/codespace/Rehabilitation_Assessment/software/data/dataset/康复评估动作/test/4/1/1.csv')
```

```python
X = pd.get_dummies(data.drop(
    'stage', axis=1), drop_first=True)
y = data['stage']

X_test = pd.get_dummies(dataTest.drop(
    'stage', axis=1), drop_first=True)
y_test = dataTest['stage']
```

```python
sns.pairplot(data, hue='stage')
# sns.pairplot(dataTest, hue='stage')
```

```python
# 训练
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# X_train, X_test, y_train, y_test = train_test_split(
#     X, y, test_size=0.3, random_state=101)
model = RandomForestClassifier(
    n_estimators=10, max_features='auto', random_state=101)
model.fit(X, y)
```

```python
# 预测
from sklearn.metrics import accuracy_score
preds = model.predict(X_test)
accuracy_score(y_test, preds)

preds
```
